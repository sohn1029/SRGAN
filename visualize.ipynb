```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
result_dataframe = pd.read_csv('./statistics/srf_4_train_results_0072e7.csv')
print(result_dataframe)
```

```python
plt.figure(figsize=(10,10)) 
plt.subplot(2,2,1)
plt.plot(result_dataframe['Epoch'], result_dataframe['Loss_D'], label = "Loss")
plt.plot(result_dataframe['Epoch'], result_dataframe['Score_D'], label = "Score")
plt.title("Discriminator")
plt.xlabel("Epoch")
plt.ylabel("Loss, Score")
plt.legend()

plt.subplot(2,2,2)
plt.plot(result_dataframe['Epoch'], result_dataframe['Loss_G'], label = "Loss")
plt.plot(result_dataframe['Epoch'], result_dataframe['Score_G'], label = "Score")
plt.xlabel("Epoch")
plt.ylabel("Loss, Score")
plt.title("Generator")

plt.subplot(2,2,3)
plt.plot(result_dataframe['Epoch'], result_dataframe['PSNR'])
plt.ylim(26,35.5)
plt.title("PSNR")
plt.xlabel("Epoch")
plt.ylabel("PSNR")

plt.subplot(2,2,4)
plt.plot(result_dataframe['Epoch'], result_dataframe['SSIM'])
plt.ylim(0.88, 1)
plt.title("SSIM")
plt.xlabel("Epoch")
plt.ylabel("SSIM")

plt.show()
```
